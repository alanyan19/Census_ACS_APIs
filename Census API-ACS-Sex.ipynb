```python
#Alan Yan
#12-19-2016
#This code requests information on the 2015 ACS's sex characteristics variables from the Census Bureau
#This code is easily adaptable to any year of the ACS available on the Census API by changing the API to 2014
# These  libraries allow the program to request data from the Census API
import requests
from urllib import parse
import json
from __future__ import division
import math
import csv
#set key
key = "blah blah blah"
#set url
base_url = "http://api.census.gov/data/2015/acs5/profile?"
#reponse format
response_format = ".json"

# The "get" key is the list of variables that the program will request
# The "for" key is the list of counties, and in this case, the program will return all counties. This part of the code can
# be adjusted to specify which counties one may desire. The list of counties can be found on the Census website.
# Finally, the key is your Census API key
search_params = {"get": "NAME,DP05_0002E,DP05_0002M,DP05_0002PE,DP05_0002PM,DP05_0003E,DP05_0003M,DP05_0003PE,DP05_0003PM,DP05_0008E,DP05_0008M,DP05_0008PE,DP05_0008PM,DP05_0009E,DP05_0009M,DP05_0009PE,DP05_0009PM,DP05_0010E,DP05_0010M,DP05_0010PE,DP05_0010PM,DP05_0011E,DP05_0011M,DP05_0011PE,DP05_0011PM,DP05_0012E,DP05_0012M,DP05_0012PE,DP05_0012PM,DP05_0013E,DP05_0013M,DP05_0013PE,DP05_0013PM,DP05_0014E,DP05_0014M,DP05_0014PE,DP05_0014PM,DP05_0015E,DP05_0015M,DP05_0015PE,DP05_0015PM,DP05_0016E,DP05_0016M,DP05_0016PE,DP05_0016PM,DP05_0017E,DP05_0017M,DP05_0017PE,DP05_0017PM",
                 "for": "county:*",
                 "key": "blah blah blah"}
# This function is a get request to the Census API. Under the current framework, all counties in the designated state is
# returned. 
def census_term(state):
    # This part sets which state to look for. Don't set the state here. One designates the state when they run the function.
    search_params['in'] = "state:" + state
    # This is where the program requests the data from the API
    r = requests.get(base_url, params=search_params)
    # I store the text in an object ie counties_list
    counties_list = r.text
    # The data object will hold the json version of the data, which allows the program to organize the data nicely
    data = json.loads(counties_list)
    # This is the list where the program stores the results of each variable
    all_counties = []
    # This iterates through each county
    for x in data:
        # This is the list of variables from the Census
        dic = {}
        dic['NAME'] = x[0].replace(",","/")
        dic['DP05_0002E'] = x[1]
        dic['DP05_0002M'] = x[2]
        dic['DP05_0002PE'] = x[3]
        dic['DP05_0002PM'] = x[4]
        dic['DP05_0003E'] = x[5]
        dic['DP05_0003M'] = x[6]
        dic['DP05_0003PE'] = x[7]
        dic['DP05_0003PM'] = x[8]
        dic['DP05_0008E'] = x[9]
        dic['DP05_0008M'] = x[10]
        dic['DP05_0008PE'] = x[11]
        dic['DP05_0008PM'] = x[12]
        dic['DP05_0009E'] = x[13]
        dic['DP05_0009M'] = x[14]
        dic['DP05_0009PE'] = x[15]
        dic['DP05_0009PM'] = x[16]
        dic['DP05_0010E'] = x[17]
        dic['DP05_0010M'] = x[18]
        dic['DP05_0010PE'] = x[19]
        dic['DP05_0010PM'] = x[20]
        dic['DP05_0011E'] = x[21]
        dic['DP05_0011M'] = x[22]
        dic['DP05_0011PE'] = x[23]
        dic['DP05_0011PM'] = x[24]
        dic['DP05_0012E'] = x[25]
        dic['DP05_0012M'] = x[26]
        dic['DP05_0012PE'] = x[27]
        dic['DP05_0012PM'] = x[28]
        dic['DP05_0013E'] = x[29]
        dic['DP05_0013M'] = x[30]
        dic['DP05_0013PE'] = x[31]
        dic['DP05_0013PM'] = x[32]
        dic['DP05_0014E'] = x[33]
        dic['DP05_0014M'] = x[34]
        dic['DP05_0014PE'] = x[35]
        dic['DP05_0014PM'] = x[36]
        dic['DP05_0015E'] = x[37]
        dic['DP05_0015M'] = x[38]
        dic['DP05_0015PE'] = x[39]
        dic['DP05_0015PM'] = x[40]
        dic['DP05_0016E'] = x[41]
        dic['DP05_0016M'] = x[42]
        dic['DP05_0016PE'] = x[43]
        dic['DP05_0016PM'] = x[44]
        dic['DP05_0017E'] = x[45]
        dic['DP05_0017M'] = x[46]
        dic['DP05_0017PE'] = x[47]
        dic['DP05_0017PM'] = x[48]
        dic['state_code'] = x[49]
        dic['county_code'] = x[50]
        # This part will add each county's name and variables to the list "all_counties"
        all_counties.append(dic)
    all_counties_key = all_counties[0]
    #This creates the file
    file_name = "ACS_2015_sex_all_states.csv"
    # This pulls from the csv library and writes the requests into a csv file
    with open(file_name, 'w') as output_file:
        dict_writer = csv.DictWriter(output_file, all_counties_key)
        dict_writer.writeheader()
        dict_writer.writerows(all_counties)
# This is my running of the function and can be changed according to one's needs. The state arg needs to be a string 
# eg if one wants to request information on California, then one would write: census_term("06")
census_term("*")
```
